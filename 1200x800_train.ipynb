```python
from googlenet import *

net = GoogLeNet(in_channel=3, num_classes=num_classes, aux_block=True).cuda()
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)

#optimizer = torch.optim.SGD(net.parameters(), lr=learning_rate, momentum=SGD_momentum)
```

```python
def validation(dataset, model, criterion):
    
    with torch.no_grad():
        model.eval()
        running_loss = 0.0
        for data in dataset:
            images, labels = data
            images = images.cuda()
            labels = labels.cuda().long()

            outputs = model(images).reshape(batch_size, num_classes)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            
    return running_loss
```

```python
times = 1
aux_loss_weight = 0.3

for epoch in range(epochs):  # loop over the dataset multiple times
    net.train()
    running_loss = 0.0
    for i, data in enumerate(mhdb_batch1):
        # get the inputs
        inputs, labels = data
        inputs = inputs.cuda()
        labels = labels.cuda().long()
        
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        aux_1, aux_2, outputs = net(inputs) #tuple: (aux1, aux2, input)
        aux_1 = aux_1.reshape(batch_size, num_classes)
        aux_2 = aux_2.reshape(batch_size, num_classes)
        outputs = outputs.reshape(batch_size, num_classes)
        
        loss1 = criterion(aux_1, labels)
        loss2 = criterion(aux_2, labels)
        loss3 = criterion(outputs, labels)
        
        loss = aux_loss_weight*loss1 + aux_loss_weight*loss2 + loss3
        loss.backward()
        ##scheduler.step()
        optimizer.step()
        
        running_loss += loss.item()
        
    if epoch % times == 0:
        print("Epoch:", epoch)
        print("Train Loss:", running_loss / times)
        val_loss = validation(val_mhdb_batch1, net, criterion)
        print("Val Loss:", val_loss)
        #print('[%d] loss: %.3f' %(epoch, running_loss / times))
        
        name = "GoogLeNet-MH"+"-Epoch-"+"%s"%epoch+"-Loss-"+"%s"%str(val_loss)[:8]+".pth"
        model_path = './saved_models/' + name
        torch.save(net.state_dict(), model_path)
        #torch.save(net, './saved_models/1200x800_%d.pkl' %epoch)
        print("Saved model:", model_path, "\n")
        
        running_loss = 0.0

print('\nFinished Training')
```
